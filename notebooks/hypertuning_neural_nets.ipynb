```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import pandas as pd
import seaborn as sns
```

## Neural Nets

```python
metric = "F1-Score"
```

### Simple RegressionNet

#### Look Back Window Size

```python
lbws = pd.read_csv("../results/tuning/simple_regression_net_lbws.csv", sep=";")
sns.barplot(x="LookBackWindowSize", y=metric, data=lbws);
```

#### Batch Size

```python
bs = pd.read_csv("../results/tuning/simple_regression_net_bs.csv", sep=";")
sns.barplot(x="BatchSize", y=metric, data=bs);
```

#### Epochs

```python
df = pd.read_csv("../results/tuning/simple_regression_net.csv", sep=";")
sns.barplot(x="Epochs", y=metric, hue="LookBackWindowSize", data=df);
```

#### Conclusion
- Set 'batch_size' to 30.
- Set 'epochs' to 10.
- Set 'look_back_window_size' to 25.

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import pandas as pd
import seaborn as sns
```

## Neural Nets

### Simple RegressionNet

#### Batch Size

```python
bs = pd.read_csv("../../results/tuning/simple_regression_net_bs.csv", sep=";")
sns.barplot(x="BatchSize", y="RMSE", data=bs);
```

#### Epochs

```python
epochs = pd.read_csv("../../results/tuning/simple_regression_net_epochs.csv", sep=";")
sns.barplot(x="Epochs", y="RMSE", data=epochs).set(yscale="log");
```

#### Look Back Window Size

```python
lbws = pd.read_csv("../../results/tuning/simple_regression_net_lbws.csv", sep=";")
sns.barplot(x="LookBackWindowSize", y="RMSE", data=lbws);
```

### Recurrent RegressionNet

```python
rec = pd.read_csv("../../results/tuning/recurrent_regression_net.csv", sep=";")
sns.barplot(x="LookBackWindowSize", y="PredictiveScore", data=rec);
```

```python
layer = pd.read_csv("../../results/tuning/recurrent_regression_net_layer.csv", sep=";")
sns.barplot(x="LookBackWindowSize", y="RMSE", hue="Layer", data=layer);
```

```python
sns.barplot(x="LookBackWindowSize", y="PredictiveScore", hue="Layer", data=layer);
```

#### Conclusion
- Set 'batch_size' to 100.
- Set 'epochs' to 10.
- Set 'look_back_window_size' to 5 or 25.
- Use GRU instead of LSTM for recurrent net.
